### Solving the quantum-mechanical problem of a particle in an infinite square potential well.

Within this notebook I will be solving the 1 Dimensional Schrodinger Equation for a particle in an infinite square well potential. I will implementing techniques such as solving differential equations using the fourth-order Runge Kutta method by defining a function and using the Newton-Raphson and Secant method to iterate values to obtain a root of the function as well as solving for boundary conditions. I will then go on to extend this code to solve the Schrodinger Equation for different potentials within the infinite square well (the infinite unsquare well) and obtain new eigen energy values corresponding to each potential.

1st potential: infinite square well potential,
2nd potential: harmonic potential within infinite square well,
3rd potential: finite square well potential within infinite square well,
4th potential: Anharmonic potential within infinite square well

```python
#Import relevant modules to be used within the code
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image # needed to import a remote image
#Command below embeds graph plots within the notebook
%matplotlib inline
```

The 1 dimensional Schrodinger Equation (eq1):
$$
\frac{-\hbar^2}{2m} \frac{d^2\psi}{dx^2} + V(x)\psi(x) = E\psi(x)
$$
Where the infinite Square Well Potential is defined as:
$$ V_x = \begin{cases} 0 & \text{if } -a < x < +a, \\ 
\infty  & \text{if } |x| > a, \end{cases} 
$$

```python
#Define parameters
m=9.1094e-31 #electron mass
hbar=1.05446e-34 #reduced Planck's constant
echarge=1.6022e-19 #electron charge
a=5e-11 #half-width of potential well in metres

#parameters for Runge Kutta Method
N=1000 # number of points for Runge-Kutta
h=2*a/N # size of Runge-Kutta steps; increment
xpoints = np.arange(-a, a, h) #array of position points with increment h
```

### Finding the ground state energy:

First I must define the Potential function, which can be altered to implement different potential wells:

```python
def V(x):
    """Potential function for the Schrodinger Equation for the infinite square well
    Inputs: (x) position of particle
    Output: (V) Potential at that position"""
    return 0.0 #potential everywhere is zero
```

We can manipulate the 1 dimensional Schrodinger Equation to form 2 first order differential equations, which are shown below (eq2 and e3 respectively):
$$
\frac{d \psi}{dx } = \phi, \\
\frac{d \phi}{dx } = \frac{2m}{\hbar^2} \left[V(x) - E \right] \psi(x),
$$
I have now defined a function which implements both 1st order differential equations and outputs an array of psi and phi values (shown below).

```python
def f(r,x,E): 
    '''Function to be solved: infinite square well potential for the Schrodinger Equation
    in the form of 2 first order ODEs.
    Inputs: (r) a vector containing psi and phi, 
            (x) the position points 
            (E) energy values
    Outputs: derivatives of psi and phi with respect to position - solutions'''
    psi = r[0] # first element of r represents psi 
    phi = r[1] # second element of r represents phi
    fpsi = phi # dpsi/dx=phi
    fphi = (2*m*(V(x)-E)*psi)/hbar**2 # dphi/dx=2m(V(x)-E)phi(x) our 1st order functions above
    #function will return dpsi/dx and dphi/dx in a vector
    return np.array([fpsi,fphi],float) # convert psi and phi to an array and ensure it's a floating point
```

Both differential equations can now be solved using the fourth order Runge Kutta Method, which iterates over values, to produce a solution of $$\frac{d \psi}{dx } and  \frac{d \phi}{dx }$$
Below I have defined the Runge Kutta function which will pass values of (x, E) through, which will then be passed through the function f(r,x,E) which I have defined above to output a psi value at each value passed through.
- Note: that r is a vector of r=r[psi,phi] which has been defined within the Runge Kutta Function
- Also within the Runge Kutta function I have not outputted the phipoints array, as I am not interested in these solutions.
- The Runge Kutta Formulas implemented here are: (eq4, 5, 6, 7, 8 respectively)
$$
k_1 = h f(x,t), \\
k_2 = h f(x+\frac{1}{2}k_1, t+\frac{1}{2}h), \\
k_3 = h f(x+\frac{1}{2}k_2, t+\frac{1}{2}h), \\
k_4 = h f(x+k_3, t+h), \\
x(t+h) = x(t) + \frac{1}{6}(k_1 + 2k_2 + 2k_3 + k_4).
$$

```python
#Define the Runge Kutta Function which solves the 1st order differential equations
def RungeKutta2d(x, E):
    '''Fourth-order Runge-Kutta for two simultaneous equations
    Inputs:(x) array of position values
           (E) array of energy values
    Outputs: [psipoints, phipoints]: solutions for psi(x) and phi(x); 
            numpy arrays of same length as xpoints'''   
    phi = 1.0 #define initial conditions
    r=np.array([0,phi]) #boundary condition; 1st component psi =0, 2nd component phi = 1.0; 2-component array
    psipoints = [] #initialise an empty array
    #Note: we do need to form a phipoints array as we do not need to output them
    for i in x:
        psipoints.append(r[0])
        k1 = h*f(r,i,E) #implementing the Runge Kutta equations
        k2 = h*f(r+0.5*k1, i+0.5*h,E)
        k3 = h*f(r+0.5*k2, i+0.5*h,E)
        k4 = h*f(r+k3, i+h,E)
        r = r + (k1 + 2*k2 + 2*k3 + k4)/6
    psipoints.append(r[0]) #append these values iterated into the empty array, r->psipoints
    return np.array([psipoints]) # convert output to numpy array with 2 rows and N columns
```

I have calculated the psi points. Now I need to solve for boundary conditions using the secant method in order to converge on a value for a root, therefore I will implement that the wavefunction is zero at x=a and x=-a (boundary conditions) and obtain a value for the energy corresponding to the ground state which satisfies this.
Secant formula: (eq9)
$$ 
x_3 = x_2 - f(x_2) \frac{x_2 - x_1}{f(x_2) - f(x_1)},
$$
Below I defined a function which implements the newton-raphson and secant method to calculate an energy (root of the secant method)

```python
def energy_gen(x,n,g1,g2):
    """Function that generates the nth excited energy state value using the secant method
    inputs: (x) position array to be passed through the Runge Kutta function
            (n) nth excited state
            (g1), (g2) guesses of ground state energy in Joules
    output: value for energy at nth excited state."""
    #guess of initial energy E1 and E2 to be inputted later
    E1 = g1*(n**2)
    E2 = g2*(n**2)
    #Use secant method to converge on right answer
    tol = echarge/1000.0 #define the tolerance
    err = 1 #initialise the error variable
    itt = 0 #initialise the iteration number
    #set up a while loop to iterate values and converge an answer
    while err > tol:
        wavefunc1 = RungeKutta2d(x,E1)[0,N] #Solve for guesses using runge kutta method to obtain wavefunction
        wavefunc2 = RungeKutta2d(x,E2)[0,N]
        E3 = E2 - wavefunc2*(E2-E1)/(wavefunc2-wavefunc1) #implement secant formula
        err = abs(E2-E1)
        # reset initial energies & iteration number for the next iteration
        E1 = E2
        E2 = E3
        itt+=1
    print("For excited state n="),n,(" Energy= {:.12} Joules taking {} iterations of the secant method.".format(E2,itt))
    return E3 #return the value of the nth excited energy state from secant method
```

### Finding the ground state wavefunction

The theoretical energy eigenvalues for an infinite square well potential are given by (eq10):
$$
E_n = \frac{\pi^2 \hbar^2 n^2}{2m (2a)^2}
$$
This equation is found when solving the Schrodinger equation for the infinite square well potential.

Due to the abitrary choice of phi, my wavefunction is not normalized. 
The normalization condition is: (eq11)
$$
\int | \psi(x) |^2 {\rm d} x = 1
$$
The easiest way to calculate the value of this integral is to use the trapezoidal rule where an integral with limits a and b can be evaluated as: (eq12)
$$
I(a,b) = h \left[ \frac{1}{2}f(a) + \frac{1}{2}f(b) +
      \sum_{k=1}^{n-1}f(x + kh) \right]
$$
Hence, I have defined a function (below) which will find the area under the integral.
However, I do not have a function to use within the trapezium rule, I have an array of x and y points instead. There function below I have adapted the trapezium rule equation (above) by finding the area of each infinitesimal trapezium corresponding to each x and y value and summing the total trapezia up to obtain a value for the total integral. 
I will then pass through the psi squared wavefuntion for ground state (the integrand) into the function 'trap' in order to find the total area under the ground state wavefunction, this gives the normalisation constant.

```python
def trap(integrand):
    """Integrate the integrand using the trapezium rule
    input: (integrand) function to be integrated
    output: value of integral"""
    #create a loop implementing trapezium rule
    integral=0.0 #initialize the integral value
    for i in range(0,len(integrand)-1):
        #calculate the area of each trapezium for the array of x and y points
        area_1=0.5*h*(integrand[i]+integrand[i+1]) #implement the trapezium rule formula
        integral=integral+area_1 #redefine integral to loop over
    return integral #return the value of the integral
```

```python
norm_constgr = trap((bestsoln[0])**2) #pass through psi squared wavefunction into 'trap'
print "The area under the curve is", norm_constgr
```

I will now compare the calculated normalised wavefunction with the theoretical ground state wavefunction, given that the normalized wavefunctions of the infinite square well potential system are given by: (eq13 and eq14 respectively)
$$
\psi_n(x)  = \frac{1}{\sqrt{a}} \cos\left(\frac{ n \pi x}{2a}\right) \qquad 
    n \ \mathrm{odd}
    $$
    $$
    \psi_n(x)  = \frac{1}{\sqrt{a}} \sin\left(\frac{ n \pi x}{2a}\right) \qquad
    n \ \mathrm{even}.
$$
Note the fact that I will use the cosine function to check this as our ground state is function is symmetrical about the centre of well and because n=1 for odd n corresponds to an even function of the form of a cosine wave. 
First I will define a function that calculates the known normalised wavefunction for a given 'n' value.

```python
#Note that every second sine and cosine curve was pi out of phase (i.e. negetive) 
#...with my calculated wavefunctions (at higher n states) so every 2nd, 3rd and 4th n we must add an odd 
#...number of pi to the phase (if an even number of pi is added to the phase it has no affect).

def theonormwf(x,n):
    """Function that will plot the theoretical normalised wavefunctions given by equations 13 and 14.
    Inputs: (x) x points array to be plotted
            (n) the nth excited state
    Output: the normalised theoretical wavefunction for the infinite square well"""
    # % modulo command shows the remainder when a division is made
    if n%2==0: #if the remainder when n is divided by 2 is 0 -> n is even -> odd function
        # // is integer divider
        theo_wf=(1/np.sqrt(a))*np.sin(n*np.pi*x/(2.0*a) + np.pi*((n//2)%2))  
    else: #n is odd -> even function
        theo_wf=(1/np.sqrt(a))*np.cos(n*np.pi*x/(2.0*a) + np.pi*((n//2)%2))
    return theo_wf #return the theoretical normalised wavefunction
```

### Finding the higher energy states:

In order to find the energy eigenvalues of high energy states, I will now need to pass through different n values through the energy_gen function, keeping the guesses to be the same as before. This is due to the fact in my energy_gen function I have accounted for the nth excited state multiplied by the guess energy. Below I have defined a function that will enable me to calculate and plot the wavefunctions for higher energy states between 1...n, it will also calculate the area under the curve and the theoretical & calculated energies for each excited state.

```python
def higherstate_wf(xpoints,xplotpoints,n_input,g1,g2):
    """Function to calculate and plot wavefunctions for higher energy states for the infinite square well potential
    for every state between 1 and n_input.
    Inputs: (xpoints) array of x points 
            (xplotpoints) array of plotted x points
            (n_input) highest nth state energy & wavefunction to be calculated
            (g1), (g2) guesses of ground state energy for secant
    Output: Comparison of calculated and theoretical energy values for states between 1 and n_input,
            Graph of the wavefunctions plotted"""
    n=1 #initialise n
    #create a loop to calculate energies of nth states and wavefunctions using previously defined functions
    while n<n_input+1: #set a limit for the loop
        energy_n=energy_gen(xpoints,n,g1,g2) #generates an energy for nth state using secant method
        bestsoln_n = RungeKutta2d(xpoints, energy_n) #solves the differential equations to obtain psi (and phi)
        print "and in eV", energy_n/echarge
        norm_wf_n = bestsoln_n[0]/np.sqrt(trap((bestsoln_n[0])**2)) #the normalised wavefunction
        print "The area under the curve is", trap(norm_wf_n**2)
        graph = plt.plot(xplotpoints, norm_wf_n, label="Calculated $\psi(x)$ n={}".format(n)) #note: syntax of label for legend
        n=n+1 #redefine new n for the loop

    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), labelspacing=0.1) #plot legend outside figure
    plt.xlabel('x (m)') 
    plt.ylabel('$\psi(x)$')
    plt.axhline(c="k",ls=":", lw=0.5) #horizontal dotted line at y=0
    return graph #function returns the graph
plt.show() #show plot
```

I also need to calculate the energy given by eq 10 (the theoretical energy) for an infinite square well potential, so that I am able to compare the theoretical energy values to the calculated energy values through the secant method

```python
def theoen(n_input):
    """Function that will calculate the theoretical energy for n states in infinite square well potential
    Input: (n_input) nth excited state
    Output: Theoretical energies between n=1...n_input"""
    n=1 # initialise n
    while n<n_input+1: #set a limit for the loop
        E_theo = (np.pi*hbar*n)**2/((2*m)*(2*a)**2) #theoretical energy formula for infinite square well
        if n==1: #use an if statement 
            print "For ground state n=",n,"Theoretical energy in Joules:", E_theo, "and in eV:", E_theo/echarge
        else:
            print "For excited state n=",n,"Theoretical energy in Joules:", E_theo, "and in eV:", E_theo/echarge
        n=n+1
    return E_theo #function returns the theoretical energy value
```

Now I will calculate and plot the normalised probability density functions for each excited state:

```python
def higherstate_prob(xpoints,xplotpoints,n_input,g1,g2):
    """Function to calculate and plot probability density functions for higher energy states 
    for the infinite square well potential; every state between 1 and n_input.
    Inputs: (xpoints) array of x points 
            (xplotpoints) array of plotted x points
            (n_input) highest nth state energy & wavefunction to be calculated
            (g1), (g2) guesses of ground state energy for secant
    Output: Comparison of calculated and theoretical energy values for states between 1 and n_input,
            Graph of the wavefunctions plotted"""
    n=1 #initialise n
    while n<n_input+1: #set a limit for the loop
        energy_n=energy_gen(xpoints,n,g1,g2) #find nth state energy
        bestsoln_n = RungeKutta2d(xpoints, energy_n) #find wavefunction using runge kutta method
        norm_prob_n = (bestsoln_n[0]/np.sqrt(trap((bestsoln_n[0])**2)))**2 #find normalisation constant and normalised wf
        graph2 = plt.plot(xplotpoints, norm_prob_n, label="Calculated $\psi^2(x)$ n={}".format(n)) #note: syntax for label of legend
        check_integral=trap(norm_prob_n)
        print "Area of integral of state n=", n, "is:", check_integral
        n=n+1 #redefine n for the loop
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), labelspacing=0.1) #plot the legend outside of graph
    plt.xlabel('x (m)')
    plt.ylabel('$\psi^2$ (x)')
    plt.axhline(c="k",ls=":", lw=0.5) #plot horizontal line y=0
    return graph2 #functions returns the graph of normalised probability density functions
plt.show() #show plot
```

### The Harmonic Potential

First I will embed a harmonic potential of the form (eq15) within the well:
$$
V(x) = V_0 \frac{x^2}{a^2}
$$
I will do this by redefining the potential function by implementing the harmonic potential

```python
Vo = 800*echarge #define the value of the potential constant Vo
```

```python
def V(x):
    """Potential function for the Schrodinger equation in the form of a harmonic potential
    Inputs: (x) position of particle
    Output: (V_x) Potential at that position"""
    return Vo*(x/a)**2 #return our potential function
```

Therefore a trial and error method was used by changing the guesses to be around the ground energy for the anharmonic potential until suitable wavefunctions were obtained.

### The Finite Square Well Potential

Now I will embed the finite potential within the infinite square well and compare this with the known wavefunctions for the finite square well, where the potential is defined as:
$$
V_x = \begin{cases} 0 & \text{if } -a/2 < x < +a/2, \\ 
    V_0  & \text{if } |x| > a/2, \end{cases} 
$$

```python
#Define the finite unsquare well potential function
#note I have still defined the function as V(x): as the runge kutta will now refer to this potential function at this point in the code
def V(x):
    """Potential function for the Schrodinger Equation for the finite square well
    Inputs: (x) position of particle
    Output: (V_x) Potential at that position"""
    if abs(x) >a/2:
        V_x = Vo #implementing our potential equation, where the potential is a constant in the well
    else: 
        V_x = 0 #everywhere else the potential is 0
    return V_x
```

The initial guesses for the finite potential within the infinite square well potential will change as the particle is under the influence of a different potential.
The guesses for the ground state energy for the finite potential within the infinite square well were made using a trial and error method, alongside knowing that the secant ultimately converges onto a root, thereby changing the guesses and narrowing down into a small range, which enabled me to find the ground state energy and higher n state energies by using this trial and error method until suitable wavefunctions and energy values (increasing as n increases) are obtained.

Note the fact that the guesses for the finite square well potential were originally taken to be either side of the ground state energy, i.e. 1.4x10^-17 and 1.5x10-17, however this produces wavefunctions which skip an energy level (i.e. n=3 became n=4 and n=4 became n=5), this was a result of the secant method iterating over the values. Therefore a trial and error method was used by changing the guesses to be around the ground energy for the anharmonic potential until suitable wavefunctions were obtained.

```python
g1fin=0.9e-17 #initial guesses for the energy in the finite potential
g2fin=1.1e-17
```

The graphs obtained above for the wavefunctions and probability density functions for the finite potential well are in agreement with the theory of the solutions to the Schrodinger equation with a finite square well potential. The main features that are in agreement are the fact that the wavefunctions and the probability density functions are in an oscillatory motion within the well (as the solutions are of the form sin(x) + cos(x), whereas outside the finite square well there is exponential decay of the wavefunction (wavefunction is of the form e$^x$ on the left side of the well and e$^{-x}$ on right side of the well), representing the fact that there is a finite probability that the particle can tunnel through the potential barrier and be found outside the finite potential well. But also the wavefunctions meet at 0 at x=+/-a which satifies the boundary conditions that there is no wavefunction outside the infinite square well, as a particle possessing infinite energy would not be possible. The wavefunctions also agree with theory as the number of maxima correspond to the nth excited state, and the probability density functions are symmmetric about the centre of the well (x=0).
